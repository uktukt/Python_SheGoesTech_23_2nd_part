```python
import pandas as pd
import sqlite3

# Connect to the database
conn = sqlite3.connect('dogs.db')
cursor = conn.cursor()

# Fetch data from the "Breeds" table
cursor.execute('''
    SELECT breed, origin, url, img FROM Breeds
''')
data = cursor.fetchall()

# Close the connection
conn.close()

# Display the data as a Pandas dataframe
df = pd.DataFrame(data, columns=['Breed', 'Origin', 'URL', 'Image'])
df.head()
```
